```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[30,25,15,10]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('line chart')
plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('bar chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[10,20,15,25]
ax[0].plot(x,y)
ax[0].set_title('line plot')
ax[1].bar(x,y)
ax[1].set_title('bar chart')
fig.suptitle('comparison of linne and bar charts')
plt.tight_layout()

plt.show()
                
```
